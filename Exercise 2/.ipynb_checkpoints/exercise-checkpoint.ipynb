```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img=cv.imread(r'spider.png')
assert img is not None

gamma=[0.2,0.8,1.2,2]
for i in gamma:
    transform=np.array([(p/255)**i*255 for p in range(0,256)]).astype(np.uint8)
    img_new=cv.LUT(img,transform)

    fig,ax=plt.subplots()
    ax.plot(transform)

    cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
    cv.imshow('Image',img)
    cv.waitKey(0)
    cv.imshow('Image',img_new)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
